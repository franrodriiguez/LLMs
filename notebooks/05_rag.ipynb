```python
import os
import pickle
import pandas as pd
import numpy as np
import faiss
import cohere
import getpass
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
PROJECT_PATH = "/content/drive/MyDrive/KeepCoding/Ai-Engineering/Practica"

DATA_PATH = os.path.join(PROJECT_PATH, "data")
os.makedirs(DATA_PATH, exist_ok=True)
```

```python
documents_path =os.path.join(DATA_PATH, "event_documents.pkl")
with open(documents_path, "rb") as f:
    documents = pickle.load(f)
```

```python
index = faiss.read_index(
    os.path.join(DATA_PATH, "faiss_index.bin")
)
```

```python
if not os.environ.get("COHERE_API_KEY"):
    os.environ["COHERE_API_KEY"] = getpass.getpass("Cohere API Key: ")

co = cohere.Client(os.environ["COHERE_API_KEY"])
```

```python
query = "Voy a organizar un tardeo universitario en Málaga para unas 1.200 personas. Dispongo de un presupuesto de marketing reducido y quiero maximizar la ocupación. ¿Qué precio de entrada recomendarías, qué estrategia de preventa utilizarías, qué canales de promoción priorizarías y qué riesgos debería tener en cuenta según eventos similares?"
```

```python
res = co.embed(
    texts=[query],
    model="embed-multilingual-v3.0",
    input_type="search_query",
    embedding_types=["float"],
)
embs = res.embeddings.float[0]
```

```python
query_vector = np.array([embs]).astype("float32")

k = 5

distances, indices = index.search(query_vector, k)
```

```python
context=""
for index, i in enumerate(indices[0],start=1):

    context+=f"Evento {index} \n"
    context+=documents[i]
    context += "\n\n"
```

```python
print(context)
```

```python
if not os.environ.get("OPENAI_API_KEY"):
    os.environ["OPENAI_API_KEY"] = getpass.getpass("OpenAI API Key: ")
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(model="gpt-5-mini",
                 temperature=0.2)


prompt = ChatPromptTemplate.from_template(
    """
Eres un asesor experto en ticketing, eventos y estrategia de venta de entradas.

Responde únicamente utilizando la información contenida en los eventos recuperados del contexto.

Normas:
- No inventes información.
- No propongas estrategias que no aparezcan o no puedan deducirse claramente del contexto.
- Si alguna información no está disponible, indícalo explícitamente.
- Justifica cada recomendación indicando el número del evento utilizado como referencia.
- Sé breve, claro y profesional.

Devuelve la respuesta con el siguiente formato:

- Precio recomendado:
- Ocupación esperada:
- Estrategia de promoción:
- Canal principal recomendado:
- Justificación (máximo 3 líneas basada en los eventos recuperados).

Contexto:
{context}

Consulta del promotor:
{query}
"""
)



rag_chain = prompt | llm | StrOutputParser()
```

```python
response = rag_chain.invoke({
    "context": context,
    "query": query
})

print(response)
```
