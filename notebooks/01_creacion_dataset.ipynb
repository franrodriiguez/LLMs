# 01 - Creación del dataset sintético

En este notebook se va a crear un dataset ficticio de eventos para alimentar el sistema RAG del proyecto AI Ticketing Advisor.

Cada fila representará un evento pasado y contendrá información útil para recomendar precio, aforo, promociones y estrategia de venta en futuros eventos.


```python
import pandas as pd
import numpy as np
import random
import os
```

```python
PROJECT_PATH = "/content/drive/MyDrive/KeepCoding/Ai-Engineering/Practica"

DATA_PATH = os.path.join(PROJECT_PATH, "data")
os.makedirs(DATA_PATH, exist_ok=True)
```

```python
random.seed(42)
np.random.seed(42)

event_types = [
    "Tardeo universitario",
    "Fiesta reggaeton",
    "Fiesta techno",
    "Concierto pop",
    "Festival urbano",
    "Evento indie",
    "Fiesta Erasmus",
    "Pool party",
    "Afterwork",
    "Noche comercial"
]

cities = ["Málaga", "Sevilla", "Granada", "Córdoba", "Madrid", "Valencia"]

venue_types = ["Discoteca", "Sala de conciertos", "Terraza", "Recinto exterior", "Beach club"]

music_genres = ["Reggaeton", "Techno", "Pop", "Urbano", "Indie", "Comercial", "Latino"]

target_audiences = [
    "Universitarios",
    "Erasmus",
    "Jóvenes 18-25",
    "Jóvenes 25-35",
    "Público general"
]

days = ["Jueves", "Viernes", "Sábado", "Domingo"]

promotion_strategies = [
    "Preventa por tramos",
    "Promoción grupos",
    "RRPP universitarios",
    "Campaña Instagram",
    "Comunidad WhatsApp",
    "Descuento early bird",
    "Sorteo de entradas",
    "Colaboración con influencers locales"
]

sales_channels = [
    "RRPP",
    "Instagram",
    "WhatsApp",
    "Web ticketing",
    "Wegow",
    "Fourvenues"
]

weather_options = ["Soleado", "Nublado", "Lluvia", "Frío", "Calor"]

business_notes_options = [
    "Buena respuesta de comunidad universitaria",
    "Ventas lentas al inicio y aceleración final",
    "Precio demasiado alto para el público objetivo",
    "Promoción por grupos funcionó muy bien",
    "Faltó campaña de urgencia en últimos días",
    "Alta conversión por RRPP",
    "Instagram generó visibilidad pero pocas ventas directas",
    "WhatsApp fue el canal más efectivo",
    "La competencia afectó a las ventas",
    "Evento con alta ocupación y buen margen"
]

rows = []

for i in range(1, 2501):
    event_type = random.choice(event_types)
    city = random.choice(cities)
    venue_type = random.choice(venue_types)
    music_genre = random.choice(music_genres)
    target_audience = random.choice(target_audiences)
    day = random.choice(days)
    promotion_strategy = random.choice(promotion_strategies)
    main_sales_channel = random.choice(sales_channels)
    weather = random.choice(weather_options)
    competing_events = random.choice([0, 1,2])
    drinks_included = random.choice([0, 1, 2])

    capacity = random.choice([300, 400, 500, 600, 800, 1000, 1200, 1500, 2000, 3000])

    prices = {
        "Tardeo universitario": 12,
        "Fiesta reggaeton": 15,
        "Fiesta techno": 18,
        "Concierto pop": 25,
        "Festival urbano": 30,
        "Evento indie": 18,
        "Fiesta Erasmus": 12,
        "Pool party": 20,
        "Afterwork": 10,
        "Noche comercial": 15
    }

    base_price = prices[event_type]

    ticket_price = base_price + random.choice([-3, -2, 0, 2, 3, 5])

    sales_start_days_before = random.choice([7, 10, 14, 21, 30, 45])

    # Reglas simples para simular ocupación
    occupancy = random.uniform(0.45, 0.95)

    if city == "Málaga":
        occupancy += 0.05

    if city == "Madrid":
        occupancy += 0.05

    if target_audience in ["Universitarios", "Erasmus"] and event_type in ["Tardeo universitario", "Fiesta Erasmus", "Fiesta reggaeton"]:
        occupancy += 0.08

    if promotion_strategy in ["Preventa por tramos", "RRPP universitarios", "Comunidad WhatsApp", "Promoción grupos"]:
        occupancy += 0.07

    if competing_events == 1:
        occupancy -= 0.10

    if weather == "Lluvia" and venue_type in ["Terraza", "Recinto exterior", "Beach club"]:
        occupancy -= 0.20

    occupancy = max(0.20, min(1.00, occupancy))

    tickets_sold = int(capacity * occupancy)
    occupancy_rate = round(tickets_sold / capacity, 2)

    if occupancy_rate >= 0.90:
        result = "Sold out o casi sold out"
    elif occupancy_rate >= 0.70:
        result = "Buen resultado"
    elif occupancy_rate >= 0.50:
        result = "Resultado medio"
    else:
        result = "Resultado flojo"

    business_notes = random.choice(business_notes_options)

    rows.append({
        "event_id": i,
        "event_type": event_type,
        "city": city,
        "venue_type": venue_type,
        "music_genre": music_genre,
        "target_audience": target_audience,
        "day_of_week": day,
        "start_time": random.choice(["16:00", "18:00", "20:00", "22:00", "00:00"]),
        "ticket_price": ticket_price,
        "capacity": capacity,
        "tickets_sold": tickets_sold,
        "occupancy_rate": occupancy_rate,
        "sales_start_days_before": sales_start_days_before,
        "promotion_strategy": promotion_strategy,
        "main_sales_channel": main_sales_channel,
        "drinks_included": drinks_included,
        "weather": weather,
        "competing_events": competing_events,
        "result": result,
        "business_notes": business_notes
    })

df_events = pd.DataFrame(rows)

df_events.head()
```
